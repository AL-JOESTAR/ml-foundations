```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('healthcare_dataset.csv', header = 0)

data = df.loc[1:, ['Age', 'Gender']]

data['Age'] = data['Age'].astype('float64')

category = ['Ringan', 'sedang', 'berbahaya' ]

data['bin_category'] = pd.qcut(data['Age'], q=3, labels=category)

data.loc[:, ['Gender', 'Age', 'bin_category']]
```

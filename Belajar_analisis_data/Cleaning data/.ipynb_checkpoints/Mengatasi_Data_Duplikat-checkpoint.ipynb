```python
import pandas as pd

data = {
    'ID' : [1,2,3,4,2],
    'Name' : ['sapri', 'udin', 'saep','ujang', 'udin']
}

df = pd.DataFrame(data)
df
```

```python
df.duplicated()
```

```python
percentage_duplicated = ( df.duplicated().sum() / len(df)) * 100
percentage_duplicated
```

```python
df_clean_duplicates = df.drop_duplicates()
df_clean_duplicates
```

```python
df.duplicated(subset='ID') # Mendeteksi duplikat berdasarkn kolom contoh kolom ID
```

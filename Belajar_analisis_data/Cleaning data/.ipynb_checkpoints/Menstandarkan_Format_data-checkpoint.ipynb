Data yang tidak konsisten dapat menyebabkan analisis menjadi tidak akurat dan menylitkan pengolahan data lebih lanjut. contohnya seperti perbedaan format pada seperti penggunaan besar/kecil ataua danya spasi tambahan

### kasus umum :
1. perbedaan huruf besar dan kecil
2. spasi tambahan
3. format tanggal yagn berbeda
4. karakter khusus
5. data yang tidak terstruktur

### Cara mengatasi inkonsistensi format dana
1. Menggunakan Fungsi standarisasi pada sofware atau toolz
   * bisa menggunakan excel (trim, upper,lower)
   * menggunakan library pandas contoh :
       * * Mengubah huruf menjadi kecil atau besar -> data['column'].str.lower()
       * * Menghapus spasi tambahan -> data['column'].str.strip()
       * * Menstandarkan format tanggal -> pd.to_datetime(data['date_column'])
2. Menggunakan regex
   * gunakan regex untuk membersihkan karakter tertentu atau menyesuaikan pola contoh menghapus karakter selain angka :
     * * data['column'].str.replace(r'[^0-9]', '')
3. membuat mapping table
   * buat referensi untuk standarisasi nilaitertentu seperti 'jalan' menjadi 'Jl'
4. Automasi dengan script ( terutama kalau dataset nya besar)
5. validasi dan verifikasi

```python
import pandas as pd

df = pd.read_csv('../Datasets/dirty_cafe_sales.csv')
sample = df.sample(30)
display(sample)
```

```python
sample['Transaction ID'] = df['Transaction ID'].str.replace('[_ ]','-', regex=True).str.lower() # str lower mereplace huruf menjadi kecil
display(sample) # contoh me replace _ menjadi -
```

## Contoh lainnya :
* str.strip() -> untuk menghapus spasi
* str.title() - > Mengubah huruf pertama menjadi kapital 'tangerang' -> 'Tangerang'

```python
# Mengisi missing value dengaen mean, median, dan modus (ini yang rata rata dipakai)

# Mean digunakan ketika : data bersifat numerik, tidak banyak outlier, dan rata rata harus merepresentasikan pusat data baik
# Contoh penggunaan mean : data nilai ujian siswa, dimana dilai hilang dan data penjualan produk dengan sedikti nilai hilang

# Median digunakna ketika : data bersifat numerik tetapi tidak terdistrubusi secara normal, dan terdapat banyak outlier yang memengaruhi rata rata
# contoh penggunaan median : data durasi kunjungan pelanggan ke situs web

# Modus digunakan ketika : data bersifat kategorikal atau nominal, kategori dengan frekuensi tertinggi dianggap paling mewakili data
# Contoh penggunaan Modus : data preferensi peloanggan (misal warna favorit, produk favorit), data demografis seperti gender, atau wilayah geografis

```

```python
# Mengisi data dengan Ffill dan Bfill
sample = df.sample(20).replace(['UNKNOWN', 'ERROR'], pd.NA) # untuk mengambil data sample sebanyak 20, dan mengubah value unknown dan eror menjadi NaN
sample
```

```python
sample['Total Spent'] = sample['Total Spent'].fillna(sample['Total Spent'].median()) # harus bertipe angka 
```

```python
sample['Total Spent'] = sample['Total Spent'].fillna(sample['Total Spent'].mean()) # harus bertipe angka 
```

```python
sample['Item'] = sample['Item'].fillna(sample['Item'].mode()) # harus bertipe huruf
```

```python
# Mengisi missing value dengan interpolasi (interpolate)
# Interpolasi digunakan untuk mengisi nilai kosong berdasarkan pola atau trend dalam data itu sendiri, cocok untuk data numerik atau data yang berurutan

# kapan digunakan : data numerik berurutan, data time series, memperkirakan nilai yanghilang berdasarkan tren
# seperti : dataset cuaca (suhu , kelembapan dll), data keuangan , data sensor
```

```python
data = {
    'Tanggal' : [
        '2026-01-01','2026-01-02','2026-01-03','2026-01-04',
        '2026-01-05','2026-01-06','2026-01-07','2026-01-08',
        '2026-01-09','2026-01-010'
    ],
    'Harga_Saham' : [
        100.0, 102.0, None, 108.0,
        None, None, 115.0, 117.0,
        None, 120.0
    ]
}

df = pd.DataFrame(data)
df
```

```python
df['Harga_Saham'] = pd.to_numeric(df['Harga_Saham'], errors='coerce')
df['Harga_Saham'] = df['Harga_Saham'].interpolate(method='linear')
df
```

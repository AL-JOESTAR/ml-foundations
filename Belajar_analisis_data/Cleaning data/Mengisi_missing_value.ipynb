```python
import pandas as pd

df = pd.read_csv('../Datasets/dirty_cafe_sales.csv')
```

```python
sample = df.sample(20).replace(['UNKNOWN', 'ERROR'], pd.NA) # untuk mengambil data sample sebanyak 20, dan mengubah value unknown dan eror menjadi NaN
```

```python
sample.fillna(0) #untuk mengisi semua kolom yang NaN
```

```python
sample['Payment Method'] = sample['Payment Method'].fillna(0) # Khusus kolomp payment method
sample
```

```python
# Mengisi data dengan Ffill dan Bfill
sample = df.head(20).replace(['UNKNOWN', 'ERROR'], pd.NA) # untuk mengambil data sample sebanyak 20, dan mengubah value unknown dan eror menjadi NaN
```

```python
sample['Payment Method'] = sample['Payment Method'].ffill # Mengisi data menggunakan ffill atau melihat data ke atas, dan bfill melihat data kebawah
sample
```

```python
sample['Payment Method'] = sample['Payment Method'].bfill()
sample
```

```python
import pandas as pd
import numpy as np
# import Datasets
df = pd.read_csv('Datasets/dirty_cafe_sales.csv')
```

```python
df.head()
```

```python
df.info() # info Datasets
```

```python
df.isna().sum() # isna = isnull (menghitungjumlah kolom yang null dalam datasets)
```

```python
df.isna().any(axis=1).sum() # cek berapa ada berapa baris yang memiliki null
```

```python
df.notna().sum() # cek jumlah kolom yang tidak null
```

```python
# Menghitung persentase missing value

missing_value = df.isna().sum()
missing_percentage = (missing_value / len(df)) * 100 # (missing value / jumlah baris) x 100 untuk menjadi percent

frame = pd.DataFrame({"missing value":missing_value, "percentage" : missing_percentage})

frame
```

```python
# ambil 20 sample data sample
sample = df.sample(20)
sample
```

```python
# Menghapus missing value
sample.dropna() # dropna(inplace=true) jika mau data langsung terhapus ke df
```

```python
sample.dropna(thresh=7) # selama setiap baris ada 8 kolom yang terisi, tidak ada dihapus
```

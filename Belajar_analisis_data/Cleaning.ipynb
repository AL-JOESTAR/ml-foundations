```python
import pandas as pd
import numpy as np
# import Datasets
df = pd.read_csv('Datasets/dirty_cafe_sales.csv')
```

```python
df.isna().sum() # isna = isnull (menghitungjumlah kolom yang null dalam datasets)
```

```python
df.isna().any(axis=1).sum() # cek berapa ada berapa baris yang memiliki null
```

```python
df.notna().sum() # cek jumlah kolom yang tidak null
```

```python
# Menghitung persentase missing value

missing_value = df.isna().sum()
missing_percentage = (missing_value / len(df)) * 100

frame = pd.DataFrame({"missing value":missing_value, "percentage" : missing_percentage})

frame
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('healthcare_dataset.csv', header = 0)

grafik = df.loc[1:20, ['Gender', 'Age', 'Medical Condition', 'Doctor']]

sns.lineplot(data=grafik, x='Gender', y='Age', label='umur dan gender')

plt.xlabel('data gender')
plt.ylabel('data umur')

plt.title('grafik umur dan gender')

plt.rcParams.update({'font.size':8})

plt.grid(True)
plt.show()
```

```python
grafik = df.loc[1:100, ['Medical Condition', 'Age']]

sns.barplot(data=grafik, x='Age', y='Medical Condition')

plt.xlabel('umur')
plt.ylabel('kondisi')
plt.title('grafik penyakit dari umur')
plt.show()
```

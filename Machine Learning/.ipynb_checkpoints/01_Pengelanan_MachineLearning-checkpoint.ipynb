## Pengenalan Machine Learning

### Supervised Machine Learning
* Terdapat dua jenis supervised learning : Classification dan Regression

* Classificiation : menekankan pada proses learning (training) untuk melakukan prediksi pada nilai discrete (kategori tertentu)
  * Contoh : Email -> Spam atau tidak spam

* Regression : menekankan pada proses learning (training) untuk melakukan prediksi pada nilai continous (Prediksi sesuatu)
  * Contoh : Prediksi harga rumah, prediksi harga saham, prediksi suhu besok

### Unsupervised Machine Learning
* Terdapat dua jenis unsupervised lerning : Clustering dan Dimensionality Reduction

* Clustering menekankan pada proses explorasi data untuk menemukan kelompok (grup) pada data berdasarkan kemiripan/kedekatan karakteristik
* Dimensionality Reduction menekankan pada proses explorasi data untuk menemukan sejumlah features yang paling berdampak terhadap response variable

## Sample Data

```python
import numpy as np
from sklearn import preprocessing

sample_data = np.array([
    [2.1, -1.9, 5.5],
    [-1.5, 2.4, 3.5],
    [0.5, -7.9, 5.6],
    [5.9, 2.3, -5.8]
])

sample_data
```

```python
sample_data.shape
```

## Binarisation

```python
sample_data
```

```python
preprocessor = preprocessing.Binarizer(threshold=0.5) # Setiap nilai yang lebih kecil atau sama dengan 0.5 akan dikonversi ke nilai 0
binarised_data = preprocessor.transform(sample_data)
binarised_data
```

## Scaling

Bertujuan untuk menghasilkan suatu data numerik yang ada dalam rentang skala tertentu

```python
sample_data
```

```python
preprocessor = preprocessing.MinMaxScaler(feature_range=(0, 1))
preprocessor.fit(sample_data)
scaled_data = preprocessor.transform(sample_data)
scaled_data
```

```python
scaled_data =preprocessor.fit_transform(sample_data)
scaled_data
```

## L1 Normalisation: Least Absolute Deviations

```python
sample_data
```

```python
l1_normalised_data = preprocessing.normalize(sample_data, norm='l1')
l1_normalised_data
```

## L2 Normalisation: Least Squares

```python
sample_data
```

```python
l2_normalised_data = preprocessing.normalize(sample_data, norm='l2')
l2_normalised_data
```

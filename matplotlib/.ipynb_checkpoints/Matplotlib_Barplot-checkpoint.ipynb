```python
import matplotlib.pyplot as plt
import numpy as np

tahun = ([2011, 2012, 2013, 2014, 2015, 2016, 2017])
jumlah = ([27, 19, 24, 13, 52, 32, 23])

fig, ax = plt.subplots()

ax.bar(tahun, jumlah)

ax.set (
    xlabel='tahun',
    ylabel='jumlah',
    title='jumlah penjualan buku pertahun'
)

plt.show()
```

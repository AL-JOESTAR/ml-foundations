```python
import matplotlib.pyplot as plt
import numpy as np

data = np.array([
    [23, 45, 12, 67, 34],
    [89, 54, 21, 76, 11],
    [43, 90, 65, 32, 18],
    [77, 28, 49, 56, 91],
    [14, 38, 72, 60, 25],
    [83, 17, 41, 95, 53],
    [66, 29, 80, 47, 36],
    [10, 58, 73, 24, 88],
    [51, 69, 33, 79, 15],
    [92, 40, 27, 64, 55]
])

fig, ax = plt.subplots()

plt.hist(data)

ax.set_title('histogram')
plt.show()
```

```python
# Contoh Customize histogram dari AI
import matplotlib.pyplot as plt
import numpy as np

# Data random dengan distribusi normal agar histogram lebih bagus
np.random.seed(42)

data = np.random.normal(
    loc=50,    # rata-rata
    scale=15,  # penyebaran data
    size=(10, 5)
)

fig, ax = plt.subplots(figsize=(8, 5))

# Flatten supaya semua data masuk ke satu histogram
plt.hist(data.flatten(), bins=10, color='skyblue', edgecolor='black')

ax.set_title('Histogram Data Random')
ax.set_xlabel('Nilai')
ax.set_ylabel('Frekuensi')

plt.grid(alpha=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

pengaruh = (['tidak berpengaruh', 'sedikit berpengaruh', 'berpengaruh', 'sangat berpengaruh'])
presentase = ([25, 30, 25, 20])

fig, ax = plt.subplots()

ax.pie(presentase, labels=pengaruh,autopct='%1.1f%%')

ax.set_title('pengaruh penggunaan aplikasi absensi online terhadap disiplin pegawai')

plt.show()
```

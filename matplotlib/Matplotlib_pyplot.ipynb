```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0,6])
ypoints = np.array([0,250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([23,6])
ypoints = np.array([40,21])

fig, ax = plt.subplots()

plt.plot(xpoints, label='x point')
plt.plot(ypoints, label='y point')

ax.set (
    xlabel= 'sumbu x',
    ylabel= 'sumbu y ',
    title='Grafik x dan y'
)

ax.legend(title='Grafik')
plt.show()
```

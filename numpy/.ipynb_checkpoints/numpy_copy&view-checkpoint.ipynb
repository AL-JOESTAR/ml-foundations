```python
# NUMPY COPY
import numpy as np

arr = np.array([1,2,3,4,5])
x = arr.copy()
arr[0] = 42

print(arr)
print(x)
```

```python
# NUMPY VIEW
arr = np.array([1,2,3,4,5])
x = arr.view()
x[0] =31

print(arr)
print(x)
```

```python
arr = np.array([1,2,3,4,5])

x = arr.copy()
y = arr.view()

print(x.base)
print(y.base)
```

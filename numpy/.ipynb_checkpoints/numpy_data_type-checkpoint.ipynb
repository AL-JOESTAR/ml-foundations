```python
import numpy as np

# Tipe data Integer
arr= np.array([1,2,3,4])

print(arr.dtype)
```

```python
# Tipe data string
arr = np.array(['apple', 'banana', 'cherry'])

print(arr.dtype)
```

```python
arr = np.array([1,2,3,4], dtype='S') # tipe data array dipaksa menjadi string byte (b)

print(arr)
print(arr.dtype) 

# S1 -> S string , 1 -> panjang karakter
```

```python
arr = np.array([1,2,3,4], dtype='i4') # Tipe data array dipaksa menjadi integer
# i4 -> i (integer), 4 (4 byte)

print(arr)
print(arr.dtype) 
```

```python
# Yang tidak bisa di convert
# arr = np.array(['a', '2', '3'], dtype='i')
```

```python
# Convert tipe data dari array yang ada

# Mengubah tipe data float menjadi integer menggunakan 'i'
arr = np.array([1.1, 2.1, 3.1])

newarr = arr.astype('i')

print(newarr)
print(newarr.dtype)
```

```python
arr = np.array([1.1, 2.1, 3.1])

# Mengubah tipe data float menjadi integer menggunakan 'int'
newarr = arr.astype('int')

print(newarr)
print(newarr.dtype)
```

```python
arr  = np.array([1,0,3])

# Mengubah tipe data integer menjadi boolean
newarr = arr.astype(bool)

print(newarr)
print(newarr.dtype)
```

```python
import numpy

# array dalam numpy
arr = numpy.array([1,2,3,4,5])

print(arr)
```

```python
# nama object array dalam numpy disebut ndarray
print(type(arr))
```

```python
# 0-D array
arr = numpy.array(41)
print(arr)
```

```python
#1-D array
arr = numpy.array([1,2,3,4,5])

print(arr)
```

```python
#2-D array
arr = numpy.array([[1,2,3],[4,5,6]])

print(arr)
```

```python
#3-D array
arr = numpy.array([[[1,2,3],[4,5,6]],
                    [[1,2,3],[4,5,6]]])

print(arr)
```

```python
# cek dimension
import numpy as np

a = np.array(41)
b = np.array([1,2,3,4,5])
c = np.array([[1,2,3],[4,5,6]])
d = np.array([[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]]])


print (a.ndim)
print (b.ndim)
print (c.ndim)
print (d.ndim)
```

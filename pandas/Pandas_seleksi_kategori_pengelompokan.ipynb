```python
# Pengelompokan Data

df = pd.DataFrame({
    "Mata Kuliah": [
        "kalkulus", "ips", "statistik", "fisika", "kimia",
        "biologi", "sejarah", "ekonomi", "bahasa", "informatika"
    ],
    "Nilai": [
        100, 90, 80, 85, 88,
        92, 75, 89, 95, 98
    ],
    "Predikat": [
        "A", "B", "C", "B", "B",
        "A", "C", "B", "A", "A"
    ]
})

kelompok = df.groupby("Predikat")[["Nilai"]].sum() # Menghitung nilai yang dikelompokan menurut predikat

print(kelompok)
```
